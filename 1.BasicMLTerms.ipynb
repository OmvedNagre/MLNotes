## Traditional Programming Vs ML

Traditinal Programming relies on explicitely rules and logic written by developers to solve problems . making it ideal for tasks with clear , predictable outcomes . 
Machine Learning enables computers to learn patterns from data and make decisions or predictions allowing them to hable complex or evolving tasks where manual rule writing is impractical . 

## Supervised vs unsupervised learning 

supervised : type of machine learning where a model is trained on data having both - inputs (features) and output (labels ) , to make prediction on new data . 

unsupervised : type of machine learning where the computer finds hidden structures in data withou being given correct ansers (no labels ).


## Classification vs Regression 

Both are supervised learning techniques 
Classification predicts categories or labels like spam/not spam, disease/no disease, etc
Regression predicts continuous values like price, temperature, sales, etc

Classification uses a decision boundary to separate data into classes, while regression fits a line through continuous data points to predict numerical values.

Decision Boundary vs Best-Fit Line : 
Best-Fit Line in Regression : 
Regression models aim to find a line or curve that best fits the continuous data points. The model minimizes prediction error using metrics(loss function) like MSE or MAE. Used when the output is numerical
Decision Boundary in Classification : 
Classification models learn a boundary that separates data into different classes. The boundary can be a straight line, curve or complex shape depending on the algorithm. Used when the output is categorical


## Semi-Supervised Learning 

Semi-supervised learning is a hybrid machine learning approach which uses both supervised and unsupervised learning. It uses a small amount of labelled data combined with a large amount of unlabelled data to train models
Semi-supervised learning is particularly valuable when acquiring labelled data is expensive or time-consuming, yet unlabelled data is plentiful and easy to collect

## Reinforcement Learning 

Learning by interacting with an environment through "trial and error" to maximize a reward . 


## Difference Between Artificial Intelligence vs Machine Learning vs Deep Learning


AI , ML , DL are no separate technologies , they are subset of one another . similar to russian nesting dolls . 

1. AI (the outer circle) 
this is the broad umbrella term for any technique that enables computers to mimic human intelligence . 

2. ML (the middle  circle) 
this is the subset of AI . it is basically the process which provides the computer to learn automatically on its own through experiences it had and accordingly withou being explicitly programmed. 

3. DL (the inner circle ) 
Deep Learning is basically a sub-part of the broader family of Machine Learning which makes use of Neural Networks(similar to the neurons working in our brain)
to mimic human brain-like behavior.
